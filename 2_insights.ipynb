```python
import pandas as pd

file_path = "historical_quiz_data.xlsx"
df = pd.read_excel(file_path, sheet_name="Sheet1")  
```

```python
# Group by 'quiz_topic' and calculate averages
topic_performance = df.groupby('quiz_topic').agg(
    avg_score=('score', 'mean'),
    avg_accuracy=('accuracy', 'mean'),
    avg_correct=('correct_answers', 'mean'),
    avg_incorrect=('incorrect_answers', 'mean')
).reset_index()

# Display the results
print(topic_performance)
```

```python
# Group by 'trophy_level' and calculate averages
trophy_performance = df.groupby('trophy_level').agg(
    avg_score=('score', 'mean'),
    avg_accuracy=('accuracy', 'mean'),
    avg_correct=('correct_answers', 'mean'),
    avg_incorrect=('incorrect_answers', 'mean')
).reset_index()

# Display the results
print(trophy_performance)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plot
sns.set(style="whitegrid")

# Plot average score by topic
plt.figure(figsize=(10, 6))
sns.barplot(x='quiz_topic', y='avg_score', data=topic_performance)
plt.title("Average Score by Topic")
plt.xlabel("Quiz Topic")
plt.ylabel("Average Score")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot average score by trophy level
plt.figure(figsize=(8, 5))
sns.barplot(x='trophy_level', y='avg_score', data=trophy_performance)
plt.title("Average Score by Trophy Level")
plt.xlabel("Trophy Level")
plt.ylabel("Average Score")
plt.show()
```
